Using the Nigeria music dataset

```python
import matplotlib as plt
import pandas as pd
import seaborn as sn
df = pd.read_csv("../Data/nigerian-songs.csv")
df.head()
```

using pandas to create scatterplot

```python
import pandas as pd
colors = ['green', 'black']
for label, color in zip(df['artist_top_genre'].unique(), colors):
    subset = df[df['artist_top_genre'] == label]
    plt.scatter(subset['popularity'], subset['danceability'], color=color, label=label)
plt.xlabel('popularity')
plt.ylabel('danceability')
plt.title('Pandas Scatterplot of Dataset')
plt.legend()
plt.show()
```
